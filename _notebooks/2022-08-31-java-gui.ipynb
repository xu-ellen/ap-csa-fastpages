```java
// implementation of GUI for Coulomb calculator

import java.awt.*;
import java.awt.event.*;
import javax.swing.*;
import java.lang.Math;

public class ForceCalculator extends JPanel implements ActionListener {
    protected JTextField textField1, textField2, textField3;
    protected JTextArea textArea;
    private final static String newline = "\n";

    private double force;   // initialize private variables
    private double q1;
    private double q2;
    private double r;
    private static double k = 8.987551787e9;
 
    public ForceCalculator() {
        super(new GridBagLayout());
 
        // Create text field for r, q1, q2
        textField1 = new JTextField(20);
        textField1.addActionListener(this);

        textField2 = new JTextField(20);
        textField2.addActionListener(this);

        textField3 = new JTextField(20);
        textField3.addActionListener(this);

        // Create text area for force
 
        textArea = new JTextArea(5, 20);
        textArea.setEditable(false);
        JScrollPane scrollPane = new JScrollPane(textArea);
 
        //Add Components to this panel.
        GridBagConstraints c = new GridBagConstraints();
        c.gridwidth = GridBagConstraints.REMAINDER;
 
        c.fill = GridBagConstraints.HORIZONTAL;
        add(textField1, c);
        add(textField2, c);
        add(textField3, c);
 
        c.fill = GridBagConstraints.BOTH;
        c.weightx = 1.0;
        c.weighty = 1.0;
        add(scrollPane, c);
    }
 
    public void actionPerformed(ActionEvent evt) {
        String r = textField1.getText();
    
        textArea.append("r=" + r + newline);
        textField1.selectAll();

        String q1 = textField2.getText();

        textArea.append("q1=" + q1 + newline);
        textField2.selectAll();

        String q2 = textField3.getText();

        textArea.append("q2=" + q2 + newline);
        textField3.selectAll();

        // calculate force if all values are entered

        if (r != null && q1 != null && q2 != null) {
            double force = (k * Double.parseDouble(q1) * Double.parseDouble(q2)) / Math.pow((Double.parseDouble(r)), 2);
            textArea.append("Force=" + force + newline);
        }
 
        //Make sure the new text is visible, even if there
        //was a selection in the text area.
        textArea.setCaretPosition(textArea.getDocument().getLength());
    }
 
    /**
     * Create the GUI and show it.  For thread safety,
     * this method should be invoked from the
     * event dispatch thread.
     */
    private static void createAndShowGUI() {
        //Create and set up the window.
        JFrame frame = new JFrame("ForceCalculator");
        frame.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
 
        //Add contents to the window.
        frame.add(new ForceCalculator());
 
        //Display the window.
        frame.pack();
        frame.setVisible(true);
    }
 
    public static void main(String[] args) {
        //Schedule a job for the event dispatch thread:
        //creating and showing this application's GUI.
        javax.swing.SwingUtilities.invokeLater(new Runnable() {
            public void run() {
                createAndShowGUI();
            }
        });
    }
}

// call main method
ForceCalculator.main(null);
```
