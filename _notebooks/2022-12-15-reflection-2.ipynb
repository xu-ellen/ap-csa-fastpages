# Unit 6: Arrays

[See Jupyter Notebook here](https://xu-ellen.github.io/ap-csa-fastpages/jupyter/cb/2022/10/25/array.html)

Takeaways:
- Arrays are a collection of data of the same type
- Arrays are indexed starting at 0
- Arrays are immutable
- Need to import java.util.Arrays to use Arrays.toString()
- Access elements through array[index]

Examples:
```java
int[] arr = {1, 2, 3, 4, 5};
System.out.println(Arrays.toString(arr));
```

# Unit 7: ArrayList

[See Jupyter Notebook here](https://xu-ellen.github.io/ap-csa-fastpages/jupyter/cb/2022/11/30/arrayList.html)

Takeaways: 
- ArrayList is a resizable array
- Primitive data types: int, double, boolean, char
- Wrapper classes: Integer, Double, Boolean, Character
- ArrayLists are mutable
- Need to import java.util.ArrayList to use ArrayList methods

Examples:
```java
ArrayList<Integer> arr = new ArrayList<Integer>();
arr.add(1);
arr.add(2);
System.out.println(arr);
```

# Unit 9: Inheritance

Takeaways:
- Inheritance is a way to reuse code
- Subclass inherits all the methods and variables from the superclass
- Subclass can override methods from the superclass
- Subclass can add new methods and variables
- PolyMorphism: the ability to call the same method on different objects and have different behaviors
